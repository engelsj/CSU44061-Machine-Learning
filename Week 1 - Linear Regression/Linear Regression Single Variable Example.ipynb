```python
import numpy as np
# data from 0 to 1 in steps of .01
Xtrain = np.arange(0,1,0.01).reshape(-1,1)
# our output has a slope of 10 with some random noise
ytrain  = 10*Xtrain + np.random.normal(0,1.0,100).reshape(-1,1)
from sklearn.linear_model import LinearRegression
# train the model
model = LinearRegression().fit(Xtrain.reshape(-1,1), ytrain.reshape(-1,1))
# intercept is the coefficient and coef is the slope
print(model.intercept_, model.coef_)
```

```python
# Plotting the predictions
ypred = model.predict(Xtrain)
import matplotlib.pyplot as plt
plt.rc('font', size = 18)
plt.rcParams['figure.constrained_layout.use'] = True
plt.scatter(Xtrain, ytrain, color = "black")
plt.plot(Xtrain, ypred, color='blue', linewidth=3)
plt.xlabel("input x")
plt.ylabel("output y")
plt.legend(["predictions","training data"])
plt.show()
```
